gender in population


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Use read_csv for .csv files
df = pd.read_csv("C:\\Users\\hp\\Downloads\\archive (2)\\age_gender.csv")

# Display the first few rows
df.head()
df.tail()
df_clean = df[['age', 'gender']].copy()
print(df)
df_clean = df_clean[df_clean['age'] <= 100]
df_clean['gender'] = df_clean['gender'].map({0: 'Male', 1: 'Female'})
df
```

```python
sns.set(style="whitegrid")  # Optional styling
plt.figure(figsize=(10, 6))
sns.histplot(data=df_clean, x='age', hue='gender', bins=30, kde=False, palette='Set2')

# Step 6: Add labels and title
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
